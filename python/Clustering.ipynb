```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist
```

```python
#Importing the dataset
data = pd.read_csv("data/Bicycle.csv", sep=",", dtype='unicode')
print("Input Data and Shape")
print(data.shape)
data.head()
#Getting the values
f1 = data['x'].values
f2 = data['y'].values
X = np.array(list(zip(f1, f2))).astype(np.float)
print(X)
```

```python
# Number of clusters
k = 4
# X coordinates of random centroids
C_x = np.random.randint(0, np.max(X), size=k)

# Y coordinates of random centroids
C_y = np.random.randint(0, np.max(X), size=k)
C = np.array(list(zip(C_x, C_y)), dtype=np.float32)

#plotting along with the Centroids
# plt.scatter(f1, f2, c='#050505', s=20)
# plt.scatter(C_x, C_y, marker="*", s=200, c='g')
# plt.xlabel("Latitude")
# plt.ylabel("Longitude")
# plt.title("Danger areas")
```

```python
#Cluster plot
kmeans = KMeans(n_clusters=k)
#Fitting the input data
kmeans = kmeans.fit(X)
#Getting the cluster labels
labels = kmeans.predict(X)
#Centroid values
centroids = kmeans.cluster_centers_
print(centroids)
for i in range(0, len(X)):
    print(labels[i])
```

```python
colors = ['r', 'g', 'b', 'y', 'c', 'm', 'o', 'w']
fig2 = plt.figure()
kx = fig2.add_subplot(111)

for i in range(k):
    points = np.array([X[j] for j in range(len(X)) if labels[j] == i])
    kx.scatter(points[:, 0], points[:, 1], s=20, cmap='rainbow')
kx.scatter(centroids[:, 0], centroids[:, 1], marker='*', s=200, c='#050505')
print("Final Centroids")
print(centroids)
plt.xlabel('Latitude')
plt.ylabel('Longitude')
plt.title('Number of clusters={}'.format(k))
```

```python
#Elbow plot
### cluster data into K=1..10 clusters #####
K_range = range(1,20)
distortions = []

for i in K_range:
    kmeanModel = KMeans(n_clusters=i)
    kmeanModel.fit(X)
    distortions.append(sum(np.min(cdist(X, kmeanModel.cluster_centers_, 'euclidean'), axis=1)) / X.shape[0])
    
fig1 = plt.figure()
ex = fig1.add_subplot(111)
ex.plot(K_range, distortions, 'b*-')
#mark the elbow
#ex.plot(K_range[2], distortions[2], marker='o', markersize=12, markeredgewidth=2, markeredgecolor='r', markerfacecolor='None')
plt.grid(True)
plt.xlim([0,10])
plt.ylim([0,0.175])
plt.xlabel('Number of clusters')
plt.ylabel('Average distortion')
plt.title('Selecting k with the Elbow Method')
```
